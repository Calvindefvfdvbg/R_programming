```r
# Date : 1 March 2025

# nested if else

Number1 <- 1000
Number2 <- 200
Number3 <- 900

# find smallest Number

# logic
# Number1 < Number2
# Number1 < Number3

# Number2 < Number1
# Number2 < Number3

# Number3 < Number1
# Number3 < Number2

if(Number1 < Number2 & Number1 < Number3){
  print("Number1 is smallest .....")
}else if(Number2 < Number1 & Number2 < Number3){
  print("Number2 is smallest .....")
}else{
  print("Number3 is smallest .....")
}
```

```r
# bill

bill <- function(price,gst){
  gst_amount <- price*gst/100
  total_price <- price+gst_amount
  cat("\n--------bill--------\n")
  cat("\nAmount = ",price,"/-")
  cat("\nGST = ",gst,"%")
  cat("\nGST Amount  = ",gst_amount,"/-")
  cat("\nTotal Price (With GST)  = ",total_price,"/-")
}
```

```r
bill(1200,18)
bill(7800,2)
bill(9000,8)
```
