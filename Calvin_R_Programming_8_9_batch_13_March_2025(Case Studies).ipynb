```r
#Date:13 March 2025
#1. Creating a Data Frame
#Create a data frame named student_data with the following columns
student_data <- data.frame(
  Name = c("Alice", "Bob", "Charlie"),
  Age = c(22, 24, 23),
  Grade = factor(c("A", "B", "A"))
)

print(student_data)
```

```r
#2. Accessing Elements in a Data Frame
#Using the student_data data frame
#Extract the Name column.
name_column<-student_data$Name
name_column<-student_data[["Name"]]
name_column<-student_data[,1]
print(name_column)
```

```r
#Access the value in the second row and third column.
value<-student_data[2,3]
print(value)
```

```r
#Extract rows where the grade is "A".
rows_with_grade_A<-student_data[student_data$Grade == "A",]
print(rows_with_grade_A)
```

```r
#3. Adding and Modifying Columns
#Add a new column Subject with values: "Math", "Science", "History".
student_data$Subject<-c("Maths","Science","History")
print(student_data)
```

```r
#Modify the Age column by adding 1 to each value.
student_data$Age<-student_data$Age+1
print(student_data)
```

```r
#4. Subsetting a Data Frame
#Display only the Name and Grade columns.
subset_data<-student_data[,c("Name","Grade")]
subset_data<-student_data[,c(1,3)]
print(subset_data)
```

```r
#Filter rows where Age is greater than 22.
filtered_data<-student_data[student_data$Age>22,]
print(filtered_data)
```

```r
#5. Summary and Structure
#Use the str() function to view the structure of the student_data.
str(student_data)
```

```r
#Use the summary() function to get a summary of the data frame.
summary(student_data)
```

```r
#6. Sorting and Reordering
#Sort the student_data by the Age column in ascending order.
sorted_data <- student_data[order(student_data$Age), ]
print(sorted_data)
```

```r
#Reorder the rows by Grade in descending order.
reorder_data <- student_data[order(student_data$Grade, decreasing = TRUE), ]
print(reorder_data)
```

```r
#7. Merging and Binding Data Frames
#Create another data frame named new_data with the following columns
new_data<-data.frame(
  Name=c("David","Eva"),
  Age=c(25,23),
  Grade=factor(c("B","A")),
  Subject=c("Chemistry","Biology")
)

print(new_data)
```

```r
#Combine student_data and new_data using rbind().
combined_data<-rbind(student_data,new_data)
print(combined_data)
```

```r
#Date:15 March 2025
#8. Removing Columns and Rows
#Remove the Subject column from the student_data.
student_data$Subject=NULL
student_data <- student_data[, !names(student_data) %in% "Subject"]
print(student_data)
```

```r
#Delete the first row from the data frame.
student_data <- student_data[-1, , drop=FALSE] #  drop=False prevents dropping dimensions
print(student_data)
```

```r
#9. Exporting and Importing Data Frames
#Save the student_data as a CSV file named student_data.csv using write.csv().
write.csv(student_data, "student_data.csv", row.names = FALSE)
print("student_data.csv has been saved successfully.")
```

```r
#Read the file back into R and store it in a variable called imported_data.
imported_data <- read.csv("student_data.csv")
print(imported_data)
```
