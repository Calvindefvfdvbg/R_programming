```r
#Bill

bill<-function (price,quantity,gst){
  total_amt<-price*quantity
  gst_amt<-total_amt*gst/100
  total_price<-total_amt+gst_amt

  cat("\n------Invoice--------\n")
  cat("\nPrice=",price,"/-")
  cat("\nQuantity",quantity,"nos")
  cat("\nTotal_Amount=",total_amt,"/-")
  cat("\nGST=",gst,"%")
  cat("\nGST_Amount=",gst_amt,"/-")
  cat("\nTotal_Price=",total_price,"/-")
}
```

```r
bill(100,6,6)
bill(1500,10,18)
```
